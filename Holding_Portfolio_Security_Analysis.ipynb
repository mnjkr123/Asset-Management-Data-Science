```python
from pyspark.sql import SparkSession

# Create a SparkSession with the JDBC driver
spark = SparkSession.builder \
    .appName("PortfolioAnalysis") \
    .getOrCreate()


# Load data from SQL Server into a DataFrame
df_holdings_cleaned = spark.read.format("jdbc") \
    .option("url", "jdbc:sqlserver://MANOJGAMING\SQLEXPRESS;databaseName=AssetAnalysis;user=admin;password=Irvine123%") \
    .option("dbtable", "Holding") \
    .load()

df_Portfolio = spark.read.format("jdbc") \
    .option("url", "jdbc:sqlserver://MANOJGAMING\SQLEXPRESS;databaseName=AssetAnalysis;user=admin;password=Irvine123%") \
    .option("dbtable", "Portfolio") \
    .load()

# Calculate the total value of holdings
df_holdings_cleaned = df_holdings_cleaned.withColumn("Value", df_holdings_cleaned["PurchasePrice"] * df_holdings_cleaned["Quantity"])

# Count the number of holdings
num_holdings = df_holdings_cleaned.count()
print(num_holdings)

# Calculate sector-wise allocation
df_sector_allocation = df_holdings_cleaned.groupBy("Sector").sum("Value")
# Show the results
df_sector_allocation.show()


from tabulate import tabulate

# Assuming 'top_holdings' is a Pandas DataFrame
top_holdings = df_holdings_cleaned.orderBy("MarketValue", ascending=False).limit(10).toPandas()

# Save the DataFrame to a CSV file
top_holdings.to_csv('top_holdings.csv', index=False)

asset_class_distribution = df_holdings_cleaned.groupBy("AssetClass").count().show()




```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from tabulate import tabulate

# Filter holdings based on a future maturity date
maturity_date = '2022-06-30'
filtered_holdings = df_holdings_cleaned.filter(df_holdings_cleaned['MaturityDate'] >= maturity_date) \
                                       .withColumnRenamed("Quantity", "TotalQuantity") \
                                       .withColumnRenamed("PurchasePrice", "TotalPurchasePrice")  \
                                       .withColumnRenamed("CouponRate", "TotalCouponRate")  \
                                       .withColumnRenamed("PropertyValue", "TotalPropertyValue")  \
                                       .withColumnRenamed("Beta", "TotalBeta") \
                                       .withColumnRenamed("MarketValue", "TotalMarketValue") \
                                       .withColumnRenamed("Sector", "TotalSector") \
                                       .withColumnRenamed("AssetClass", "TotalAssetClass")


# Check if the filtered DataFrame is empty
if filtered_holdings.count() == 0:
    print("No holdings found with maturity date on or after", maturity_date)
else:
    # Join with Portfolio DataFrame to get Portfolioname
    joined_holdings = filtered_holdings.join(df_Portfolio, 'PortfolioID', 'inner')

    # Prepare features
    feature_columns = ['TotalQuantity', 'TotalPurchasePrice', 'TotalCouponRate', 'TotalPropertyValue', 
                       'TotalBeta']
    assembler = VectorAssembler(inputCols=feature_columns, outputCol='features')
    df_assembled = assembler.transform(joined_holdings)
    
    
    # Train a linear regression model
    lr = LinearRegression(featuresCol='features', labelCol='MarketValue')
    lr_model = lr.fit(df_assembled)

    # Predict market value
    predictions = lr_model.transform(df_assembled)

    # Group by Portfolioname, Sector, and AssetClass, and calculate the sum of predicted market value
    grouped_predictions = predictions.groupBy('Portfolioname', 'TotalSector', 'TotalAssetClass') \
        .agg({'prediction': 'sum'}) \
        .withColumnRenamed('sum(prediction)', 'PredictedMarketValue')

    # Show the results
    grouped_predictions.show()
    

import pandas as pd

# Convert the Spark DataFrame to a Pandas DataFrame
grouped_predictions_pd = grouped_predictions.toPandas()

# Save the Pandas DataFrame to a CSV file
grouped_predictions_pd.to_csv('grouped_predictions.csv', index=False)





```

```python
from pyspark.ml.feature import StringIndexer
from pyspark.ml.classification import RandomForestClassifier

# Convert Sector to numerical labels
indexer = StringIndexer(inputCol='Sector', outputCol='SectorIndex')
df_indexed = indexer.fit(df_holdings_cleaned).transform(df_holdings_cleaned)

# Prepare features
feature_columns = ['Quantity', 'PurchasePrice', 'CouponRate', 'PropertyValue', 'Beta', 'SectorIndex']
assembler = VectorAssembler(inputCols=feature_columns, outputCol='features')
df_assembled = assembler.transform(df_indexed)

# Train a random forest classifier
rf = RandomForestClassifier(featuresCol='features', labelCol='SectorIndex')
rf_model = rf.fit(df_assembled)

# Predict sector
predictions = rf_model.transform(df_assembled)

# Show the results
predictions.show()

import pandas as pd

# Convert the Spark DataFrame to a Pandas DataFrame
predictions_pd = predictions.toPandas()

# Save the Pandas DataFrame to a CSV file
predictions_pd.to_csv('predictions.csv', index=False)
```
